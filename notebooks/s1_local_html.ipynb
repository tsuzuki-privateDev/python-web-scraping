```python
from bs4 import BeautifulSoup
import os

path = r"/home/tadatsugu/python-web-scraping/data/sample.html"
# BeautifulSoupオブジェクトを作成
with open(path, 'r', encoding='utf-8') as f:
    html = f.read()
    soup = BeautifulSoup(html, "html.parser")

print(soup.title.text)
print(soup.find("h1").text)
print(len(soup.find_all("li", class_="item")))
```

```python
items = soup.find_all("li", class_="item")
results = []

for item in items:
    name = item.find('span', class_="name").text
    price = int(item.find('span', class_="price").text)
    item_dict = {
        "name": name,
        "price": price
    }
    results.append(item_dict)

print(results)
```

```python
from pathlib import Path
import csv

out_path = "outputs/items.csv"
# フォルダを作成
Path("outputs").mkdir(exist_ok=True)

def write_items_csv(items, out_path):
    with open(out_path, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=["name", "price"])
        writer.writeheader()
        for row in items:
            writer.writerow(row)

write_items_csv(results, out_path)
```
